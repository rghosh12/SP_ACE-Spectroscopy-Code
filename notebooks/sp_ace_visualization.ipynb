````python
from pathlib import Path

# Repository root (parent of notebooks/)
ROOT = Path.cwd().resolve()
if ROOT.name == "notebooks":
    ROOT = ROOT.parent

import sys
if str(ROOT) not in sys.path:
    sys.path.insert(0, str(ROOT))

%matplotlib inline
from fortran_tools.visualization import load_fortran_run_outputs, plot_run_dashboard

# After running Fortran, set ``PAR`` to your ``space.par`` (outputs share the same stem).
# The optional demo cell below sets ``PAR`` for a tiny in-repo fixture.
PAR = ROOT / "tests" / "fixtures" / "_notebook_demo.par"
OUT = ROOT / "notebooks" / "_dashboard_out.png"
````

````python
# Optional: build tiny demo outputs (same stem as ``space.par``) without running Fortran.
# For real work, **skip this cell** and set ``PAR`` in the **first code cell** to your ``myrun.par`` after ``space`` finishes.

import shutil

demo_par = ROOT / "tests" / "fixtures" / "_notebook_demo.par"
demo_par.write_text("# placeholder\n", encoding="utf-8")
stem_base = demo_par.parent / demo_par.stem  # e.g. .../ _notebook_demo  (no .par)
shutil.copy(ROOT / "tests" / "fixtures" / "example_model_snippet.dat", f"{stem_base}_model.dat")
shutil.copy(ROOT / "tests" / "fixtures" / "minimal_TGM_ABD.dat", f"{stem_base}_TGM_ABD.dat")
shutil.copy(ROOT / "tests" / "fixtures" / "minimal_ew_meas.dat", f"{stem_base}_ew_meas.dat")
PAR = demo_par
print("Demo outputs written for", PAR)
````
